```python
# Introduction:
# The Netflix TV Shows Dataset
# 
# In this data analysis project, we will explore the Netflix TV Shows dataset. This dataset contains information about various TV shows available on the Netflix platform, including details about their directors, cast, release years, ratings, and more. Our goal is to gain insights into the content offered by Netflix, analyze audience preferences, and visualize trends in show types, genres, and more. We will start by loading the dataset and performing initial data exploration and data cleaning to prepare it for in-depth analysis and visualization.

# Step 1: Importing Libraries
# We start by importing essential libraries for data analysis and visualization.
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Step 2: Loading the Dataset
# We load the Netflix TV shows dataset from a CSV file into a pandas DataFrame.

df = pd.read_csv('netflix.csv')
```

```python
# Step 3: Initial Data Overview
# To understand the dataset, we display the first few rows, data types, and summary statistics.
df.head()  # Display the first few rows
df.info()  # Display data types and non-null counts
df.describe()  # Generate summary statistics
```

```python
# Step 4: Handling Missing Data
# We address missing data in several columns by replacing it with suitable placeholders.
df.isnull().sum()  # Check for missing data
df.dropna()  # Remove rows with missing data
df['director'].fillna('Not Available', inplace=True)
```

```python
df['cast'].fillna('Not Available', inplace=True)
```

```python
df['date_added'].fillna('Not Available', inplace=True)
```

```python
df['rating'].fillna('Not Rated', inplace=True)
```

```python
# Step 5: Distribution of Ratings (Bar Chart)
# We visualize the distribution of show ratings using a bar chart to understand audience preferences.
sns.countplot(data=df, x='rating')
plt.title('Distribution of Ratings')
plt.xlabel('Rating')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
# Step 6: Distribution of Show Types (Pie Chart)
# We create a pie chart to visualize the distribution of show types (TV Shows and Movies).
show_type_counts = df['type'].value_counts()
plt.figure(figsize=(6, 6))
plt.pie(show_type_counts, labels=show_type_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Show Types')
plt.show()
```

```python
# Step 7: Distribution of Release Years (Bar Plot)
# We use a bar plot to show the distribution of release years for the shows.
release_year_counts = df['release_year'].value_counts().sort_index()
plt.figure(figsize=(10, 6))
sns.barplot(x=release_year_counts.index, y=release_year_counts.values)
plt.title('Distribution of Release Years')
plt.xlabel('Release Year')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
# Step 8: Distribution of Show Durations (Histogram)
# We create a histogram to visualize the distribution of show durations.

plt.figure(figsize=(8, 6))
sns.histplot(df['duration'], bins=20, kde=True)
plt.title('Distribution of Show Durations')
plt.xlabel('Duration (minutes)')
plt.ylabel('Frequency')
plt.show()
```

```python
# Step 9: Stacked Plot of Distribution of Show Types by Ratings
# The stacked bar chart illustrates the distribution of show types within different ratings categories.

rating_show_type = df.groupby(['rating', 'type']).size().unstack()
rating_show_type.plot(kind='bar', stacked=True, figsize=(10, 6))
plt.title('Distribution of Show Types by Ratings')
plt.xlabel('Rating')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.legend(title='Type')
plt.show()
```

```python
# Step 10: Heatmap of Genre Popularity Over the Years
# A heatmap is used to visualize how the popularity of different genres has changed over time.

genre_release = df.pivot_table(index='release_year', columns='genres', aggfunc='size', fill_value=0)
plt.figure(figsize=(12, 8))
sns.heatmap(genre_release, cmap='viridis')
plt.title('Genre Popularity Over the Years')
plt.xlabel('Genre')
plt.ylabel('Release Year')
```

```python
# Step 11: Box Plot of Show Duration by Types
# We create a box plot to show the distribution of show durations by show types.
plt.figure(figsize=(8, 6))
sns.boxplot(data=df, x='type', y='duration')
plt.title('Box Plot of Show Durations by Type')
plt.xlabel('Type')
plt.ylabel('Duration (minutes)')
plt.show()
```
